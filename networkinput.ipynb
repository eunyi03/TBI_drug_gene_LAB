```python
subacute_df = pd.read_csv("/data1/project/eunyi/tbi/drug_gene/bowht_subacute_deseq2_gene.csv", index_col=0)
subacute_df
```

```python
subacute_df_up = subacute_df[(subacute_df['Subacute padj'] < 0.22)  & (subacute_df['TBI Subacute vs SControl Log2 fold change'] > 0.4)]
subacute_df_up
```

```python
subacute_df_down = subacute_df[(subacute_df['Subacute padj'] < 0.0002)  & (subacute_df['TBI Subacute vs SControl Log2 fold change'] < -1.3)]
subacute_df_down
```

```python
up_deg = subacute_df[(subacute_df['Subacute padj'] < 0.22) & (subacute_df['TBI Subacute vs SControl Log2 fold change'] > 0.4)]
down_deg = subacute_df[(subacute_df['Subacute padj'] < 0.0002) & (subacute_df['TBI Subacute vs SControl Log2 fold change'] < -1.3)]
output_sub_up = pd.read_csv('0.adjusted_subacute_up.txt', sep='\t', header=None, names=['gene', 'score'])
output_sub_down = pd.read_csv('0.adjusted_subacute_down.txt', sep='\t', header=None, names=['gene', 'score'])
merged_up = up_deg.merge(output_sub_up, left_on='gene', right_on='gene')
merged_down = down_deg.merge(output_sub_down, left_on='gene', right_on='gene')

print(merged_up)
print(merged_down)
```

```python
merged_up.sort_values(by='score', ascending=False)
```
